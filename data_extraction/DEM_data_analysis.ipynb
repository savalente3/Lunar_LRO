```python
import matplotlib.pyplot as plt
import seaborn as sns

from LRO_data_class import LunarDataset
```

```python
ds = LunarDataset().DEMLunarData
ds.info()
```

```python
display(ds.describe())

print('Shape: ', ds.shape)
print('Duplicates: ', ds.duplicated().sum())
print('Null values')
print(ds.isna().sum())
```

## Image quality

```python
sns.histplot(ds.values.flatten()[::1000], bins=50, kde=True)
plt.axvline(x=0, color='red', linestyle='--', label='Datum (0m)')

plt.xlabel('Elevation')
plt.ylabel('Count')
plt.title('DEM Elevation Distribution (SLDEM2015)')
plt.legend()
```
