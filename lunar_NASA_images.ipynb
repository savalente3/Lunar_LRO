```python
import requests
import os
```

```python
headers = {"Authorization": f"Bearer {os.environ['NASA_EARTHDATA_TOKEN']}"}

response = requests.get(
    "https://cmr.earthdata.nasa.gov/search/granules.json",
    headers=headers,
    params={
        "short_name": "LRO_LROC_CDR",
        "bounding_box": "-180,-90,180,90",
        "page_size": 10
    }
)

print(response.json())
```
